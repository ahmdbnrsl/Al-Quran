````python
import numpy as np

range_hour, range_value = 5, 100

data = []
for i in range(2500):
    hour = np.random.randint(0, range_hour)
    value = 0
    if 4 < hour <= 5:
        value = np.random.randint(90, 100)
    elif 3 < hour <= 4:
        value = np.random.randint(70, 92)
    elif 2 < hour <= 3:
        value = np.random.randint(65, 75)
    elif 1 < hour <= 2:
        value = np.random.randint(55, 66)
    elif 0 <= hour <= 1:
        value = np.random.randint(0, 55)

    data.append([hour, value])

real_data = np.array(data)
weight, bias = 5, 5
n = len(real_data)
learning_rate = 0.051

for i in range(1500):
    total_squared = sum(((weight * x + bias) - y)**2 for x, y in real_data)
    MSE = 1/n * total_squared

    d_w = 1/n * sum(2 * ((weight * x + bias) - y) * x for x, y in real_data)
    d_b = 1/n * sum(2 * ((weight * x + bias) - y) for x, y in real_data)

    weight = weight - learning_rate * d_w
    bias = bias - learning_rate * d_b

def predict(x):
    return weight * x + bias

predict(4)
````

$$
